### Making plots with plotnine
1. Call `ggplot()` function and give it a DataFrame
2. Then, we define what aspects of the plot will correspond to which variables in the DataFrame with the `aes()` function.
3. Specify a geometry. We will primarliy use scatter plots, boxplots and density plots.

```python
import plotnine as p9

(p9.ggplot([pandas Dataframe]) + p9.aes(x='variable to put on X-axis',
                                        y='variable to put on Y-axis',
                                        color='variable') + p9.geom_point())
```

#### Scatter plot
- `geom_point()`
- With scatter plot each row in the dataframe will correspond to one point, and each axis will correspond to one variable. Additionally we pass sex to the color argument to color the points according to sex.

```python
import plotnine as p9
import pandas as pd

df = pd.DataFrame(data= {'Sex':['Male','Male','Female','Female'],
                         'Height (cm)':[183, 179, 160, 172],
                         'Weight (kg)':[82, 75.1, 50, 58.7]})

print(p9.ggplot(df) + p9.aes(x='Height (cm)', y='Weight (kg)', color='sex') + p9.geom_point())
```

#### Boxplot
- `geom_boxplot()`

```python
import plotnine as p9
import pandas as pd

df = pd.DataFrame(data= {'Sex':['Male','Male','Female','Female'],
                         'Height (cm)':[183, 179, 160, 172],
                         'Weight (kg)':[82, 75.1, 50, 58.7]})

print(p9.ggplot(df) + p9.aes(x='sex', y='Height', fill='sex') + p9.geom_boxplot())
```

#### Density plot
- `geom_density()`

```python
import plotnine as p9
import pandas as pd

df = pd.DataFrame(data= {'Sex':['Male','Male','Male','Male','Male','Male','Female','Female','Female','Female','Female','Female'],
                         'Height (cm)':[183, 179, 160, 172,190,181,175, 160,165,158,154,170]})

print(p9.ggplot(df) + p9.aes(x='Height', fill='sex') + p9.geom_density(alpha=0.5))
```
